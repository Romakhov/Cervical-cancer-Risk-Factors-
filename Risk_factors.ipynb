```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

%matplotlib notebook
```

```python
data = pd.read_csv('risk_factors_cervical_cancer.csv')
```

```python
data=data.replace('\?','0',regex=True)
data.astype(float)
data.head()
```

Определить биопсию

```python
X = data.iloc[:,1:32]
Y = data['Biopsy']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=123, test_size=0.3)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
```

```python
#обучение моделей

svm = SVC(gamma='auto').fit(x_train, y_train) 
DecisionTree = tree.DecisionTreeClassifier().fit(x_train, y_train) 
NaiveB = GaussianNB().fit(x_train, y_train) 
LogisticRegression = LogisticRegression().fit(x_train, y_train) 
```

```python
# Предсказание

pred_svm = svm.predict(x_test)
pred_DecisionTree = DecisionTree.predict(x_test)
pred_NaiveB = NaiveB.predict(x_test)
pred_LogisticRegression = LogisticRegression.predict(x_test)
```

```python
# Оценка эффективности

print('Точность SVM: ', accuracy_score(y_test, pred_svm))
print('Точность дерева решений: ', accuracy_score(y_test, pred_DecisionTree))
print('Точность наивного байесовского алгоритма: ', accuracy_score(y_test, pred_NaiveB))
print('Точность логистической регрессии: ', accuracy_score(y_test, pred_LogisticRegression))
```

Наивысшую точность показали SVM и Логистическая регрессия
